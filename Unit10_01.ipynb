# 우리 동네 인구 구조를 산점도로 나타내기

## 꺾은선 그래프로 표현하기

```python
import csv
f = open('gender.csv')
data = csv.reader(f)

m = []
f = []
```

```python
name = input('궁금한 동네를 입력해주세요. : ')
for row in data:
    if name in row[0]:
        for i in range(3, 104):
            m.append(int(row[i]))
            f.append(int(row[i+103]))
        break
```

```python
import matplotlib.pyplot as plt
plt.plot(m, label='male')
plt.plot(f, label='female')
plt.legend()
plt.show()
```

남여 인구의 차이가 더 잘 보임. 특히 10~30대에서는 남자가 더 많고 60대 이상에서는 여자가 더 많다.

## 막대그래프로 표현하기

연령대별로 남성 인구에서 여성 인구를 뺀 값을 가시적으로 보여주려고 한다.

```python
import csv
f = open('gender.csv')
data = csv.reader(f)
result = []
name = input('궁금한 동네를 입력해주세요. : ')

for row in data:
    if name in row[0]:
        for i in range(3, 104):
            result.append(int(row[i])-int(row[i+103]))
        break
#for i in range(3, 104):
#    result.append(m[i]-f[i])    로 표현해도 된다.
```

```python
import matplotlib.pyplot as plt
plt.bar(range(101), result)
plt.show()
```

now we can see the gap between male and female population.

## 산점도로 표현하기

다양한 색상과 크기의 점을 통해 남녀 연령대별 인구수를 나타내보자! scatter() 함수를 써보자.

```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')
plt.scatter([1, 2, 3, 4], [10, 30, 20, 40])   #bar() 함수처럼 x축, y축
plt.show()
```

```python
# scatter() 함수를 통해 버블 차트로 표현하기
#size 추가하기
import matplotlib.pyplot as plt
plt.scatter([1, 2, 3, 4], [10, 30, 20, 40], s=[100, 200, 250, 300])
plt.show()
```

```python
# color 속성 추가하기
import matplotlib.pyplot as plt
plt.scatter([1, 2, 3, 4], [10, 30, 20, 40], s=[30, 60, 90, 120], c=['r', 'b', 'g', 'gold'])
```

```python
# 컬러바 추가하기
#colorbar() 사용
import matplotlib.pyplot as plt
plt.scatter([1, 2, 3, 4], [10, 30, 20, 40], s=[30, 60, 90, 120], c=range(4))
plt.colorbar()
plt.show()
```

```python
# 컬러바에 사용될 색상의 종류를 정하기
import matplotlib.pyplot as plt
plt.scatter([1, 2, 3, 4], [10, 30, 20, 40], s=[30, 60, 90, 120], c=range(4), cmap='jet') # 'jet' color map 사용
plt.colorbar()
plt.show()
```

## 랜덤 함수를 이용해 무작위의 점 100개를 산점도 위에 나타내보기

```python
import matplotlib.pyplot as plt
import random
x = []
y = []
size = []

for i in range(100):
    x.append(random.randint(50, 100))
    y.append(random.randint(50, 100))
    size.append(random.randint(10, 100))
plt.scatter(x, y, s=size, c=size, cmap='coolwarm', alpha=0.7)   # 사이즈에 따른 컬러 구분, 70% 투명도 설정
plt.colorbar()
plt.show()
```

```python
import matplotlib.pyplot as plt
import random
x = []
y = []
size = []

for i in range(100):
    x.append(random.randint(50, 100))
    y.append(random.randint(50, 100))
    size.append(random.randint(10, 100))
plt.scatter(x, y, s=size, c=x, cmap='jet')   # 사이즈에 따른 컬러 구분, 70% 투명도 설정
plt.colorbar()
plt.show()
```
