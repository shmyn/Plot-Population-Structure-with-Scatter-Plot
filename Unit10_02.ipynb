# 제주도의 연령대별 성별 비율을 산점도로 표현하기
x축에 여자 인구수를, y축에 남자 인구수를 입력해서 여자 인구가 더 많으면 x축 쪽으로, 남자 인구가 더 많으면 y축 쪽으로 치우치게 설계한다.

```python
import csv
f = open('gender.csv')
data = csv.reader(f)
m = []
f = []
name = input("궁금한 동네를 입력해주세요. : ")
for row in data:
    if name in row[0]:
        for i in range(101):
            m.append(int(row[i+3]))
            f.append(int(row[i+106]))
        break
```

```python
import matplotlib.pyplot as plt
plt.scatter(m, f)
plt.show()
```

```python
import matplotlib.pyplot as plt  #굳이 다시 언급해보았다
plt.scatter(m, f, c=range(101), alpha=0.5, cmap='jet')
plt.colorbar()
plt.plot(range(max(m)), range(max(m)), 'g') #추세선
plt.show()
```

여기서 점의 사이즈까지 표현해보겠다. 남자 인구수와 여자 인구수를 합치면 된다.
# 마지막 !

```python
import csv
import math
f = open('gender.csv')
data = csv.reader(f)

m = []
f = []

name = input("궁금한 동네를 입력해주세요. : ")

size = []
for row in data:
    if name in row[0]:
        for i in range(101):
            m.append(int(row[i+3]))
            f.append(int(row[i+106]))
            size.append(math.sqrt(m[i]+f[i]))  #math.sqrt를 추가하지 않으면 엄청난 크기의 동그라미들을 보게 된다..b
        break

import matplotlib.pyplot as plt
plt.style.use('ggplot')
plt.rc('font', family = 'Malgun Gothic')
plt.title(name+' 지역의 성별 인구 그래프')
plt.scatter(m, f, s=size, c=range(101), alpha=0.5, cmap='jet')
plt.colorbar()
plt.plot(range(max(m)), range(max(m)), 'g')
plt.xlabel('남성 인구수')
plt.ylabel('여성 인구수')
plt.show()
```

+sqrt() 함수를 추가하지 않으면

```python
import csv
import math
f = open('gender.csv')
data = csv.reader(f)

m = []
f = []

name = input("궁금한 동네를 입력해주세요. : ")

size = []
for row in data:
    if name in row[0]:
        for i in range(101):
            m.append(int(row[i+3]))
            f.append(int(row[i+106]))
            size.append(m[i]+f[i])  #sqrt() 생략
        break

import matplotlib.pyplot as plt
plt.style.use('ggplot')
plt.rc('font', family = 'Malgun Gothic')
plt.title(name+' 지역의 성별 인구 그래프')
plt.scatter(m, f, s=size, c=range(101), alpha=0.5, cmap='jet')
plt.colorbar()
plt.plot(range(max(m)), range(max(m)), 'g')
plt.xlabel('남성 인구수')
plt.ylabel('여성 인구수')
plt.show()
```
